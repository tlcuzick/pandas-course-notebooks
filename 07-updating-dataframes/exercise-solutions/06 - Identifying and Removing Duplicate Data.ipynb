## Setup

```python
import pandas as pd
```

```python
orders_path = 'YOUR PATH HERE'
```

```python
orders = pd.read_csv(orders_path)
orders.head()
```

## Use the `duplicated` method to identify any rows from this dataset that are duplicated across all five columns:

```python
dupes = orders.duplicated()
orders[dupes]
```

```python
# verify that one of the duplicate rows identified is actually a dupe
orders[orders['OrderID'] == 124]
```

## Use the `drop_duplicates` method to remove any rows from this dataset that are duplicated across all five columns:

```python
orders.drop_duplicates(inplace=True)
orders[orders['OrderID'] == 124]
```

## Identify any rows from this dataset that are duplicated across just the "CustomerID", "OrderDate", and "SubTotal" columns:

```python
dupes = orders.duplicated(subset=['CustomerID', 'OrderDate', 'SubTotal'])
orders[dupes]
```

```python
# verify that one of the duplicate rows identified is actually a dupe
orders[(orders['CustomerID'] == 1562) & (orders['SubTotal'] == 37.47)]
```

## Remove any rows that are duplicated across just the "CustomerID", "OrderDate", and "SubTotal" columns:

```python
orders.drop_duplicates(subset=['CustomerID', 'OrderDate', 'SubTotal'], inplace=True)
orders[(orders['CustomerID'] == 1562) & (orders['SubTotal'] == 37.47)]
```

## Reset the index of the DataFrame:

```python
orders.reset_index(inplace=True, drop=True)
orders.head()
```
