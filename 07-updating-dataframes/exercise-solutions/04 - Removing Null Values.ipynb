## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.info()
```

## Remove all rows from the UFO sightings dataset with a `Null` value in the "state" column:

```python
no_null_states = sightings.dropna(subset='state')
no_null_states.info()
```

## How many rows were removed from the DataFrame?

```python
len(sightings) - len(no_null_states)
```

**5,797**

## Remove all rows from the UFO sightings dataset with a `Null` value in either the "state" OR "country" columns:

```python
sightings.dropna(subset=['state','country'], inplace=True)
sightings.info()
```

## How many rows in the original DataFrame had `Null` values in the "country" column but not the "state" column?

```python
len(no_null_states) - len(sightings)
```

**6,414**
