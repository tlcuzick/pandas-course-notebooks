## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
```

## Capture the current average of "duration (seconds)" in a variable:

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
avg_duration = sightings['duration (seconds)'].mean()
avg_duration / 60
```

## Update the values in the "duration (seconds)" field such that any values greater than 3600 are converted to exactly 3600:

```python
gt_3600 = sightings['duration (seconds)'] > 3600
sightings.loc[gt_3600, 'duration (seconds)'] = 3600
```

## Compare the average of the updated "duration (seconds)" field to the original average:

```python
sightings['duration (seconds)'].mean() / 60
```
