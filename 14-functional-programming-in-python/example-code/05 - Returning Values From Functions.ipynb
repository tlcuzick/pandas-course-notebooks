```python
def percent_change(prev, curr):
  return ((curr - prev) / prev) * 100
```

```python
percent_change(5, 7)
```

```python
delta = percent_change(5, 1)
```

```python
delta
```

```python
def house_rater(price, bathrooms, grade, condition):
  if price >= 400000:
    return 'Too expensive!'
  elif (bathrooms - int(bathrooms)) in(0.25, 0.75):
    return 'Stupid number of bathrooms!'
  elif (grade > 7) and (condition > 3):
    return "I'm interested!"
  else:
    return "Not interested:("
```

```python
rating = house_rater(333000, 3, 8, 4)
```
