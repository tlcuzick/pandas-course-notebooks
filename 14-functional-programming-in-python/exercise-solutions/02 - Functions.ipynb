## Create a function called `calculate_calories_burned` that estimates the number of calories burned during exercise. The function should take the type of activity and duration (in minutes) as inputs and **return**:

* 10 calories per minute for running

* 8 calories per minute for swimming

* 7 calories per minute for cycling

* 5 calories per minute for any other activity

```python
def calculate_calories_burned(activity, duration):
    if activity == "running":
        return duration * 10
    elif activity == "swimming":
        return duration * 8
    elif activity == "cycling":
        return duration * 7
    else:
        return duration * 5
```

```python
cals = calculate_calories_burned('running', 30)
```

## Create a function called `determine_winner` that determines the winner of a rock-paper-scissors game. The function should take the choices of two players as input (either "rock", "paper", or "scissors") and **print**:

* "It's a tie!" if both players choose the same

* "Player 1 wins!" if player 1 wins

* "Player 2 wins!" if player 2 wins

*(For reference, rock beats scissors, paper beats rock, and scissors beats paper.)*

```python
def determine_winner(player1, player2):
    if player1 == player2:
        print("It's a tie!")
    elif player1 == "rock" and player2 == "scissors":
        print("Player 1 wins!")
    elif player1 == "scissors" and player2 == "paper":
        print("Player 1 wins!")
    elif player1 == "paper" and player2 == "rock":
        print("Player 1 wins!")
    else:
        print("Player 2 wins!")
```

```python
determine_winner("rock", "paper")
```

## Write a function called `calculate_ticket_price` that calculates the price of a movie ticket based on the customer's age and the day of the week. Assume a base price of $12. The function should return:

* 80% of the base price for children (under 12) and seniors (65 and older)

* 90% of the base price for Wednesday showings

* The full base price for all other cases

```python
def calculate_ticket_price(age, day_of_week):
    base_price = 12
    if age < 12 or age >= 65:
        return base_price * 0.8
    elif day_of_week == "Wednesday":
        return base_price * 0.9
    else:
        return base_price
```

```python
price = calculate_ticket_price(14, "Wednesday")
```
