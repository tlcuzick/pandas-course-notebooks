## Given variables `temperature` (in Celsius) and `oxygen_percentage`, write an `if` statement to determine if it's safe for a space explorer to land on a planet. If the temperature is between -20 and 50 (inclusive) AND the oxygen percentage is at least 19.5%, print "Safe to land". Otherwise, print "Unsafe for landing".

```python
temperature = 20
oxygen = 18
```

```python
if (temperature >= -20) and (temperature <= 50) and (oxygen >= 19.5):
    print("Safe to land")
else:
    print("Unsafe for landing")
```

## Given variables `mood` (1-10) and `is_rainy` ("yes" or "no"), write an `if` statement to make mood-based music recommendations. Print "Upbeat pop" if mood is 7 or higher and it's not rainy, "Melancholic tunes" if mood is 3 or lower, and "Neutral playlist" otherwise.

```python
mood = 8
rainy = 'yes'
```

```python
if (mood >= 7) and (rainy != "yes"):
    print("Upbeat pop")
elif mood <= 3:
    print("Melancholic tunes")
else:
    print("Neutral playlist")
```

## Given variables `plant_type` ("succulent", "tropical", or "herb"), `days_since_watered`, and `sunlight_exposure` ("low", "medium", or "high"), write an `if` statement that gives plant care advice. Print "Don't water yet!" for succulents watered less than 14 days ago, "Move to a brighter spot!" for tropical plants with low sunlight, "Time to water!" for herbs not watered in more than 3 days, and "Your plant is okay for now:)" otherwise.

```python
plant_type = "tropical"
days_since_watered = 11
sunlight = "low"
```

```python
if plant_type == "succulent" and days_since_watered < 14:
    print("Don't water yet!")
elif plant_type == "tropical" and sunlight == "low":
    print("Move to a brighter spot!")
elif plant_type == "herb" and days_since_watered > 3:
    print("Time to water!")
else:
    print("Your plant is okay for now.")
```
