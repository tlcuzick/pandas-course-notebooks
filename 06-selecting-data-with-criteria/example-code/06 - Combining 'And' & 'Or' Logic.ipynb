```python
import pandas as pd
```

```python
house_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(house_path)
```

```python
house_sales[house_sales['waterfront'] > 0].head()
```

```python
waterfront = house_sales['waterfront'] == 1
nice_view = house_sales['view'] > 0
multi_story = house_sales['floors'] > 1
```

```python
life_goals = house_sales[waterfront | nice_view & multi_story]
life_goals.head(20)
```

```python
price = house_sales['price'] <= 300000
basement = house_sales['sqft_basement'] > 0
floors = house_sales['floors'] > 1
```

```python
multi_level = house_sales[price & basement | floors]
multi_level.head(20)
```

```python
multi_level = house_sales[price & (basement | floors)]
multi_level.head(20)
```

```python
basement_or_floors = basement | floors
```

```python
multi_level = house_sales[price & basement_or_floors]
multi_level.head(10)
```

```python
multi_level = house_sales[(house_sales['price'] <= 300000) & ((house_sales['sqft_basement'] > 0) | (house_sales['floors'] > 1))]
multi_level.head(10)
```
