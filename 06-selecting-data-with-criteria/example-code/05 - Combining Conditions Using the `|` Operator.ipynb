```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head()
```

```python
sqft_living = house_sales['sqft_living'] >= 4000
sqft_lot = house_sales['sqft_lot'] >= 12000
bedrooms = house_sales['bedrooms'] >= 5
waterfront = house_sales['waterfront'] == 1
```

```python
fancy_houses = house_sales[sqft_living | sqft_lot | bedrooms | waterfront]
fancy_houses.head(20)
```
