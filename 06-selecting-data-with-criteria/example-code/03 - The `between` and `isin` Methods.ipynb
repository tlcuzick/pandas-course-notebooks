```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head()
```

```python
between_6_n_10 = house_sales['bedrooms'].between(6, 10)
between_6_n_10_df = house_sales[between_6_n_10]
```

```python
between_6_n_10_df.sort_values('bedrooms', inplace=True)
between_6_n_10_df.head()
```
