## WITHOUT the use of a `~` operator, fetch all records from the UFOs dataset that do NOT meet the following set of criteria:

* "country" = 'us'...

* OR "latitude" is between 33 and 38 degrees...

* OR "shape" is one of 'light, 'circle' or 'disk'

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
countries = sightings['country'] != 'us'
latitude_gt_38 = sightings['latitude'] > 38
latitude_lt_33 = sightings['latitude'] < 33
null_latitude = sightings['latitude'].isna()
shape_nt_eq_light = sightings['shape'] != 'light'
shape_nt_eq_circle = sightings['shape'] != 'circle'
shape_nt_eq_disk = sightings['shape'] != 'disk'

subset_1 = sightings[countries & (latitude_gt_38 | latitude_lt_33 | null_latitude) & shape_nt_eq_light & shape_nt_eq_circle & shape_nt_eq_disk]
subset_1.head(10)
```

```python
len(subset_1)
```

## Refactor the above code using the `~` operator:

```python
us = sightings['country'] == 'us'
latitudes = sightings['latitude'].between(33,38)
shapes = sightings['shape'].isin(['light','circle','disk'])

subset_1a = sightings[~(us | latitudes | shapes)]
subset_1a.head(10)
```

```python
len(subset_1a)
```

## Define the logical inverse of the solution code from the exercises after the previous lecture ("Combining 'And' & 'Or' Logic"):

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
shapes = sightings['shape'].isin(["rectangle", "cylinder","cone"])
states = sightings['state'].isin(['ky','wv','tn'])
duration = sightings['duration (seconds)'] < 5

states_or_duration = states | duration

subset_2 = sightings[~(shapes & states_or_duration)]
subset_2.head(10)
```
