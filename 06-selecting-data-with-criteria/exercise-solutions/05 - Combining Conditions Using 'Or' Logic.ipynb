## Use "or" logic to generate a filtered subset of UFO sightings that meet the following criteria:

* The sightings must have occurred in one of the following states: Washington (abbreviated "wa"), or Oregon ("or").

* OR, the sightings must have occurred between 33 and 38 degrees latitude.

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
wa_or = sightings['state'].isin(['wa','or'])
bt_33_38_lat = sightings['latitude'].between(33,38)

subset_1 = sightings[wa_or | bt_33_38_lat]
subset_1.head(10)
```

## How many sightings met these criteria?

```python
len(subset_1)
```

**26,686**

## Use "or" logic to generate a filtered subset of UFO sightings that meet the following criteria:

* The sightings must occurred either above 66.5 degrees latitude OR below -66.5 degrees latitude.

* OR, the sightings must have occurred in state of Colorado (abbreviated "co").

```python
# It's totally fine (and probably preferable) if you chose to assign each boolean series to a variable BTW; I just opted for the condensed approach in this case to keep you familiar with that syntax.

subset_2 = sightings[(sightings['latitude'] > 66.5) | (sightings['latitude'] < -66.5) | (sightings['state'] == 'co')]
subset_2.head(10)
```

## How many sightings met these criteria?

```python
len(subset_2)
```

**1524**
