## Setup

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## What is the highest latitude of a sighting identified with a "shape" of cylinder?

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
cylinders = sightings[sightings['shape'] == 'cylinder']
cylinders['latitude'].max()
```

**71.3875 degrees**

## How does the average duration of sightings in the United States compare to the median duration of sightings in Australia?

```python
sightings['duration (minutes)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce') / 60
```

```python
us_sightings = sightings[sightings['country'] == 'us']
us_sightings['duration (minutes)'].mean()
```

```python
au_sightings = sightings[sightings['country'] == 'au']
au_sightings['duration (minutes)'].mean()
```

**Average duration of sightings in US: 96.67 minutes**

**Average duration of sightings in Australia: 63.44 minutes**
