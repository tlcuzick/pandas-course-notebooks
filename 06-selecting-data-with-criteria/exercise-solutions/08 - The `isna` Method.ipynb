## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Fetch all records with a null "shape" value in the UFO sightings dataset, into their own distinct DataFrame:

```python
null_shapes = sightings[sightings['shape'].isna()]
null_shapes.head()
```

## What is the most frequently occurring "state" among those sightings with a null "shape"?

```python
null_shapes.describe(include='all')
```

**"wa" (Washington)**

## Identify any values in the "latitude" column of the UFO sightings dataset that can't be converted to numbers:

```python
null_latitudes = pd.to_numeric(sightings['latitude'], errors='coerce').isna()

sightings[null_latitudes]['latitude']
```

**33q.200088**

## How many values in the "datetime" columns of the UFO sightings dataset can't be converted to dates?


```python
null_datetimes = pd.to_datetime(sightings['datetime'], errors='coerce').isna()

len(sightings[null_datetimes])
```

**694**
