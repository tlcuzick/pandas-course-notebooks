## How does the median square footage of houses built in the 1930s compare to houses built in the 1940s?

```python
house_path = 'YOUR PATH HERE'
house_sales = pd.read_csv(house_path)
```

```python
sales_1930s = house_sales[house_sales['yr_built'].between(1930, 1939)]
sales_1930s['sqft_living'].median()
```

```python
sales_1940s = house_sales[house_sales['yr_built'].between(1940, 1949)]
sales_1940s['sqft_living'].median()
```

**Median square footage of houses built in 1930s: 1,635**

**Median square footage of houses built in 1940s: 1,370**

## How does the total number of UFO sightings in the cities of Miami, Orlando, and Jacksonville (combined) compare to the total number of sightings in Seattle, Portland, and Vancouver combined?

```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
```

```python
moj_sightings = sightings[sightings['city'].isin(['miami', 'orlando', 'jacksonville'])]
len(moj_sightings)
```

```python
spv_sightings = sightings[sightings['city'].isin(['seattle', 'portland', 'vancouver'])]
len(spv_sightings)
```

**Total number of sightings in Miami, Orlando, and Jacksonville: 631**

**Total number of sightings in Seattle, Portland, and Vancouver: 1,035**
