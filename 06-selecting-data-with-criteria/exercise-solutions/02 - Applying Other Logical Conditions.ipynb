## Setup

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## How does the average duration of sightings in the United States compare to the average duration of sightings everywhere else?

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
us_sightings = sightings[sightings['country'] == 'us']
us_sightings['duration (seconds)'].mean() / 60
```

```python
non_us_sightings = sightings[sightings['country'] != 'us']
non_us_sightings['duration (seconds)'].mean() / 60
```

**Average duration of US sightings: 96.7 minutes**

**Average duration of non-US sightings: 379.7 minutes**

## What is the median latitude among sightings less than or equal to 10 seconds in duration?

```python
short_sightings = sightings[sightings['duration (seconds)'] <= 10]
short_sightings['latitude'].median()
```
