## Use "and" logic to generate a filtered subset of UFO sightings that meet the following criteria:

* The sightings must have occurred in one of the following states: Kentucky (abbreviated "ky"), Tennessee ("tn"), or West Virginia ("wv").

* AND, the sightings must have been 60 seconds or less in duration.

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
ky_tn_wv = sightings['state'].isin(['ky','tn','wv'])
lte_60_sec = sightings['duration (seconds)'] <= 60

subset_1 = sightings[ky_tn_wv & lte_60_sec]
subset_1.head(10)
```

## How many sightings met these criteria?

```python
len(subset_1)
```

**908**

## Use "and" logic to generate a filtered subset of UFO sightings that meet the following criteria:

* The sightings must have been over 10 minutes in duration.

* AND, the sightings must have occurred in the United States (abbreviated "us").

* AND, the sightings must have occurred between 33 and 38 degrees latitude.

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
gt_10_min = sightings['duration (seconds)'] > 600
in_us = sightings['country'] == 'us'
bt_33_38_lat = sightings['latitude'].between(33,38)

subset_2 = sightings[gt_10_min & in_us & bt_33_38_lat]
subset_2.head(10)
```

## How many sightings met these criteria?

```python
len(subset_2)
```

**4449**
