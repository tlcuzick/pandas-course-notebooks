```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Generate a filtered subset of UFO sightings that meet the following criteria:

* The sightings must have been classified as having one of the following shapes: "rectangle", "cylinder", or "cone". **This condition must be met.**

* The sightings must have occurred in one of the following states: Kentucky (abbreviated "ky"), Tennessee ("tn"), or West Virginia ("wv")...
OR, the sightings must have been less than 5 seconds in duration. **Only ONE of these two conditions must be met.**

*Use a separate series for each column-level condition, and then combine them when you index the DataFrame.*

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'],errors='coerce')
```

```python
shapes = sightings['shape'].isin(["rectangle", "cylinder","cone"])
states = sightings['state'].isin(['ky','wv','tn'])
duration = sightings['duration (seconds)'] < 5

subset_1 = sightings[shapes & (states | duration)]
subset_1.head(10)
```

## Refactor your code from Exercise 1 as follows:

Combine any groups of related conditions (e.g., the either/or conditions) into their own consolidated series, then use these when you index the DataFrame.

```python
shapes = sightings['shape'].isin(["rectangle", "cylinder","cone"])
states = sightings['state'].isin(['ky','wv','tn'])
duration = sightings['duration (seconds)'] < 5

states_or_duration = states | duration

subset_2 = sightings[shapes & states_or_duration]
subset_2.head(10)
```
