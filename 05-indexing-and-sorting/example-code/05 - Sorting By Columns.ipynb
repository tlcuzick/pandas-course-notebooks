```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings['duration (seconds)'] = pd.to_numeric(ufo_sightings['duration (seconds)'],errors='coerce')
ufo_sightings['datetime'] = pd.to_datetime(ufo_sightings['datetime'], errors='coerce')
ufo_sightings.head()
```

```python
ufo_sightings.sort_values('datetime', inplace=True)
ufo_sightings.head()
```

```python
oldest_100 = ufo_sightings.iloc[0:100]
oldest_100.describe(include='all')
```

```python
ufo_sightings.sort_values('datetime', inplace=True, ascending=False)
ufo_sightings.head()
```

```python
oldest_100.sort_values(['country', 'state', 'city', 'duration (seconds)'], inplace=True)
oldest_100.head(10)
```
