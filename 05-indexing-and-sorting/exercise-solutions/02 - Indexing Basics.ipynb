## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

## Convert the "datetime" column to a datetime data type:

```python
ufo_sightings['datetime'] = pd.to_datetime(ufo_sightings['datetime'], errors='coerce')
```

## Set the index of the DataFrame to be the "datetime" column, while retaining the "datetime" column in the data itself:

```python
ufo_sightings.set_index('datetime',inplace=True,drop=False)
ufo_sightings.head()
```
