## Setup

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head()
```

## Fetch the records associated with the 1000 smallest houses in the house sales dataset (by square feet of living space) into a DataFrame:

```python
smallest_1000 = house_sales.sort_values('sqft_living').iloc[0:1000]
smallest_1000.describe()
```

## What is the average price of these 1000 houses?

**$285,978.60**

## What is the median "yr_built" value across these 1000 houses?

**1943**

## Now answer the same questions as above, but for the 1000 largest houses (again, by square feet of living space) in the dataset:

```python
largest_1000 = house_sales.sort_values('sqft_living', ascending=False).iloc[0:1000]
largest_1000.describe()
```
