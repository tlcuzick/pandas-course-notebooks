## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
sightings = pd.read_csv(ufos_path)
```

```python
orders_path = 'YOUR PATH HERE'
orders = pd.read_csv(orders_path)
```

## In the UFO sightings dataset, create a "posting delay - years" field:

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings['date posted'] = pd.to_datetime(sightings['date posted'],errors='coerce')
```

```python
sightings['posting delay - years'] = (sightings['date posted'] - sightings['datetime']).dt.days / 365.25
sightings.head()
```

## In the orders dataset, create an "EstimatedShippingTime" field - positioned immediately following the "OrderTimestamp" field - based on the order timestamp plus 12 hours:

```python
orders['OrderTimestamp'] = pd.to_datetime(orders['OrderTimestamp'], errors='coerce')
```

```python
est = orders['OrderTimestamp'] + pd.Timedelta(hours=12)
orders.insert(4, 'EstimatedShippingTime', est)
orders.head()
```
