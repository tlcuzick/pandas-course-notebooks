## Compare the percentage of **all** sightings that occurred in the U.S. before 9/10/1993, versus on or after 9/10/1993:

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
```

```python
before_x_files = sightings[sightings['datetime'] < '09/10/1993']
len(before_x_files[before_x_files['country'] == 'us']) / len(before_x_files)
```

```python
on_after_x_files = sightings[sightings['datetime'] >= '09/10/1993']
len(on_after_x_files[on_after_x_files['country'] == 'us']) / len(on_after_x_files)
```

**Before 9/10/1993, 78.4% of sightings in the dataset occurred in the U.S.**

**On and after 9/10/1993, 81.4% of sightings in the U.S. occurred in the dataset.**

## Compare the number of **overall** sightings in the 5 years prior to 9/10/1993 to the number of overall sightings in the 5 years on or after 9/10/1993:

```python
five_yrs_prior = sightings[sightings['datetime'].between('09/10/1988', '09/10/1993')]
len(five_yrs_prior)
```
