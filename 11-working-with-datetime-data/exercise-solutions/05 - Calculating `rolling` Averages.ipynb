## Setup

```python
import pandas as pd
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

## Using the house sales dataset, create a breakout of number of homes sold AND average home sale price by year and month.

```python
house_sales['date'] = pd.to_datetime(house_sales['date'], errors='coerce')
```

```python
house_sales['month'] = house_sales['date'].dt.month
house_sales['year'] = house_sales['date'].dt.year
```

```python
year_month_summary = house_sales.groupby(['year', 'month']).agg(homes_sold=('id','count'), avg_price=('price', 'mean'))
year_month_summary
```

## Add a second column to this view which displays the **weighted** average of home sale price for the previous three months, relative to a given year/month in the view.

```python
num_sold_x_avg_price = year_month_summary['homes_sold'] * year_month_summary['avg_price']
```

```python
rolling_num_sold_x_avg_price = num_sold_x_avg_price.rolling(window=3).sum()
rolling_num_sold = year_month_summary['homes_sold'].rolling(window=3).sum()
```

```python
year_month_summary['weighted_3_month_avg_price'] = rolling_num_sold_x_avg_price / rolling_num_sold
year_month_summary
```
