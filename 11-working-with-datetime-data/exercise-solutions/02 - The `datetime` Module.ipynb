## Break out the average sighting latitude by quarter. Which quarters had the highest and lowest average sighting latitudes? Can you come up with a hypothesis to (potentially) explain these results?

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'], errors='coerce')
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
sightings['quarter'] = sightings['datetime'].dt.quarter
```

```python
sightings.groupby('quarter')['latitude'].mean()
```

**The first quarter (Januray through March) had the lowest average sighting latitude, while the third quarter (July through September) had the highest.**

**In the northern hemisphere, where the vast majority of sightings occurred, weather tends to be especially cold - and days especially short - at higher latitudes, while the opposite is true in the third quarter. One possible hypothesis which could explain the results, is that more people in the northen hemisphere, *across more different latitudes*, simply tend to be outside more (making UFO sightings possible) in the third quarter versus the first.**

## Summarize the count of sightings by day of month. Which two days of the month had most total sightings? Can you come up with a plausible hypothesis to explain *these* results?

```python
sightings['day of month'] = sightings['datetime'].dt.day
sightings['day of month'].value_counts()
```

**It seems noteworthy that the 1st and the 15th days of the month have historically had more sightings by a considerable margin than the other days of the month.**

**Equally worth noting is that companies that pay their employees bi-monthly, often do so on the 1st and the 15th of the month. It's not inconceivable that people who have just gotten paid are more likely to go out at night (and potentially party, drink, etc.), making UFO sightings more plausible across a wider portion of the day.**

## What percentage of overall sightings occur between 7pm and 7am?

```python
sighting_hour = sightings['datetime'].dt.hour
sightings_1900_to_700 = sightings[(sighting_hour >= 19) | (sighting_hour < 7)]
len(sightings_1900_to_700) / len(sightings)
```

**Around 76% of sightings occurred between 7m and 7am.**
