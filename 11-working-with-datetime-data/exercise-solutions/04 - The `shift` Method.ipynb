## Setup

```python
import pandas as pd
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

## Create a summarized view that compares average home sale prices by year and month, to average sale prices in the previous year/month period:

```python
house_sales['date'] = pd.to_datetime(house_sales['date'], errors='coerce')
```

```python
house_sales['month'] = house_sales['date'].dt.month
house_sales['year'] = house_sales['date'].dt.year
```

```python
avg_price_by_year_month = house_sales.groupby(['year', 'month'])['price'].mean()
avg_price_by_year_month = avg_price_by_year_month.to_frame()
```

```python
avg_price_by_year_month['price - prev month'] = avg_price_by_year_month['price'].shift(1)
avg_price_by_year_month
```

## Enhance this view by adding a "% delta" column that displays the percentage increase/decrease from the previous month's average sale price to the current month's:

```python
avg_price_by_year_month['% delta'] = ((avg_price_by_year_month['price'] - avg_price_by_year_month['price - prev month'])/ avg_price_by_year_month['price - prev month']) * 100
avg_price_by_year_month
```
