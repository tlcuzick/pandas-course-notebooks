```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings.head()
```

```python
import datetime as dt
```

```python
sightings.insert(1, 'year', sightings['datetime'].dt.year)
sightings.insert(2, 'month', sightings['datetime'].dt.month)
sightings.head()
```

```python
sightings[sightings['month'] == 10]
```

```python
sightings['duration (minutes)'] = pd.to_numeric(sightings['duration (seconds)'],errors='coerce') / 60
sightings['latitude'] = pd.to_numeric(sightings['latitude'],errors='coerce')
```

```python
sightings.groupby('month')[['duration (minutes)', 'latitude']].mean()
```

```python
sightings['weekday'] = sightings['datetime'].dt.day_name()
sightings['weekday'].value_counts()
```

```python
sightings[sightings['datetime'].dt.dayofweek >= 5]
```

```python
sightings['datetime'].dt.hour.value_counts()
```

```python
sightings[sightings['datetime'].dt.hour.between(0,4)]
```
