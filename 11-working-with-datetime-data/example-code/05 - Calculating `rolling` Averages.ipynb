```python
import pandas as pd
import datetime as dt
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
```

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings['duration (minutes)'] = pd.to_numeric(sightings['duration (seconds)'],errors='coerce') / 60
sightings['year'] = sightings['datetime'].dt.year
```

```python
sightings_by_year = sightings.groupby('year').agg(num_sightings=('date posted', 'count'), avg_duration=('duration (minutes)','mean'))
sightings_by_year.tail(10)
```

```python
sightings_by_year['num_sightings'].rolling(window=5).mean()
```

```python
num_sightings_rolling_5 = sightings_by_year['num_sightings'].rolling(window=5).mean()
sightings_by_year.insert(1, 'num_sightings_rolling_5', num_sightings_rolling_5)
sightings_by_year.tail(10)
```

```python
avg_duration_rolling_3 = sightings_by_year['avg_duration'].rolling(window=3).mean()
sightings_by_year.insert(3, 'avg_duration_rolling_3', avg_duration_rolling_3)
sightings_by_year.tail(10)
```
