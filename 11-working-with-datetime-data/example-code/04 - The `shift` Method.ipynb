```python
import datetime as dt
```

```python
sightings = pd.read_csv(ufos_path)
```

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
```

```python
sightings['year'] = sightings['datetime'].dt.year
```

```python
sightings_by_year = sightings.groupby('year')['date posted'].count()
sightings_by_year.tail(10)
```

```python
sightings_by_year = sightings_by_year.to_frame(name='# sightings')
sightings_by_year.tail(10)
```

```python
prev_year = sightings_by_year['# sightings'].shift(1)
sightings_by_year['# sightings - prev year'] = prev_year
sightings_by_year.tail(10)
```

```python
sightings_by_year.to_excel('Sightings By Year.xlsx',"Data")
```

```python
sightings_by_year['YoY Delta'] = sightings_by_year['# sightings'] - sightings_by_year['# sightings - prev year']
sightings_by_year.tail(10)
```
