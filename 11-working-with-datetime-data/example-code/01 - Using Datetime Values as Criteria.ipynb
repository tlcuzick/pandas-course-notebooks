```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

```python
sightings.info()
```

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
```

```python
sightings[sightings['datetime'] >= '2014-04-05'].sort_values('datetime').head(10)
```

```python
sightings[sightings['datetime'] >= '2014-04-05 16:20:00'].sort_values('datetime').head(10)
```

```python
sightings[sightings['datetime'] < '1930-01-01'].sort_values('datetime').head(10)
```

```python
date_mask = sightings['datetime'].between('2014-04-05 11:11:00', '2014-04-05 17:55:00')
sightings[date_mask].sort_values('datetime').head(10)
```
