## Use a text cell to explain the formula for converting Fahrenheit to Celsius. Then, in a code cell, create a variable for a Fahrenheit temperature and convert it to Celsius.

The formula for converting Fahrenheit to Celsius is: subtract 32 from the Fahrenheit temperature, then multiply the result by 5/9.

```python
fahrenheit = 88
celsius = (fahrenheit - 32) * (5/9)
celsius
```
