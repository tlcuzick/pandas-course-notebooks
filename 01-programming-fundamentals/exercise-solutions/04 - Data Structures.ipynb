## Create a dictionary - assigned to a variable named "movie_info" with the following structure:

Keys: Movie titles

Values: Another dictionary containing:

    "director": Director's name (string)

    "year": Year released (integer)

    "genre": The movie's genre (string)

Include information for at least two movies. Then display the director of one of the movies by accessing it from the nested dictionary.

```python
movie_info = {
    "Desperado": {
        "director": "Robert Rodriguez",
        "year": 1995,
        "genre": "Action"
    },
    "Army of Darkness": {
        "director": "Sam Raimi",
        "year": 1992,
        "genre": "Comedy/Horror"
    }
}
```

```python
movie_info["Desperado"]["director"]
```

## Add a new key-value pair to each of the nested dictionaries you created in the previous exercise, whose value consists of an abbreviated list  of the actors/actresses starring in the movie. The corresponding key should be called "cast".

```python
movie_info["Desperado"]["cast"] = ["Antonio Banderas", "Salma Hayek"]
movie_info["Army of Darkness"]["cast"] = ["Bruce Campbell", "Embeth Davidtz"]
```

## Add a new actor/actress to the "cast" list for one of the movies, and display the updated ("movie_info") dictionary.

```python
movie_info["Desperado"]["cast"].append("Steve Buscemi")
movie_info
```
