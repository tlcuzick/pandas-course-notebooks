```python
import pandas as pd
import datetime as dt
%matplotlib inline
```

```python
house_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(house_path)
houses.head()
```

```python
houses.plot()
```

```python
houses['price'].plot()
```

```python
houses['date'] = pd.to_datetime(houses['date'],errors='coerce')
```

```python
price = houses['price']
price.index = houses['date']
price.plot()
```

```python
houses['year'] = houses['date'].dt.year
houses['month'] = houses['date'].dt.month
```

```python
houses_grouped = houses.groupby(['year', 'month'])['price'].mean()
houses_grouped
```

```python
houses_grouped.plot()
```
