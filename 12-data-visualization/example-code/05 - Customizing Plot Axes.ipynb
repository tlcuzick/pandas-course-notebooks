```python
import pandas as pd
import datetime as dt
%matplotlib inline
```

```python
house_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(house_path)
```

```python
houses['date'] = pd.to_datetime(houses['date'],errors='coerce')
houses['year'] = houses['date'].dt.year
houses['month'] = houses['date'].dt.month
houses_grouped = houses.groupby(['year', 'month'])['price'].mean()
```

```python
houses_grouped.plot(color='#08cc0b', title='King County House Sales', xlabel='Year and Month', ylabel='Sale Price')
```

```python
line_plot = houses_grouped.plot(color='#08cc0b', title='King County House Sales', xlabel='Year and Month', ylabel='Sale Price')
line_plot.set_yticks([0,100000,200000,300000,400000,500000,600000]);
line_plot.set_yticklabels(['0','100K','200K','300K','400K','500K','600K']);
```
