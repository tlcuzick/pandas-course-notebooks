```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
%matplotlib inline
```

```python
house_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(house_path)
houses.head()
```

```python
houses['bedrooms'].head(100).plot(kind='bar')
```

```python
num_bedrooms = houses['bedrooms'].value_counts()
num_bedrooms
```

```python
num_bedrooms.plot(kind='bar')
```

```python
num_bedrooms.sort_index(inplace=True)
num_bedrooms.plot(kind='bar')
```

```python
price_by_grade = houses.groupby('grade')['price'].mean()
price_by_grade
```

```python
price_by_grade.plot(kind='bar')
```
