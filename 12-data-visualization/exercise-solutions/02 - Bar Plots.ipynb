## Setup

```python
import pandas as pd
%matplotlib inline
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Using a **bar** plot, visualize the total number/count of sightings by month:

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings['month'] = sightings['datetime'].dt.month
```

```python
num_sightings_by_month = sightings['month'].value_counts().sort_index()
num_sightings_by_month.plot(kind='bar')
```

## Does this specific breakout seem to be better-suited for a bar plot versus a line plot?

**Yes, bar plots are generally better-suited to visualize categorical breakouts like this one.**

## Using a bar plot, visualize the average sighting *latitude* by sighting *shape*:

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
avg_latitude_by_shape = sightings.groupby('shape')['latitude'].mean()
avg_latitude_by_shape.plot(kind='bar')
```

## What two shapes stand out for having unusually high average latitudes?

**Sightings with a shape of "crescent" or "pyramid" seem to have occurred at a significantly higher latitude, on average, than other sightings.**

## Using a bar plot, visualize the total number/count of sightings by *country*:

```python
sightings['country'].value_counts().sort_index().plot(kind='bar') # All on one line of code this time!
```

## Now change the previous plot to a **pie** plot/chart:

```python
sightings['country'].value_counts().sort_index().plot(kind='pie')
```

## Does this visualization work as well or better than the bar plot?

**In my oh-so-very humble opinion, it does a slightly better job of conveying the proportionality of the sighting count across the five countries represented in our dataset.**

**While it has become quite fashionable in data viz circles to malign pie charts, they are actually a perfectly fine - and in some cases, marginally preferable - alternative to bar charts for visualizing proportionality, especially if you're plotting a very small number of categories (ideally, 5 or fewer).**
