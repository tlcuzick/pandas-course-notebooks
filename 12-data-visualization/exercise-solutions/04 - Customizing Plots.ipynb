```python
import pandas as pd
%matplotlib inline
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Recreate the line plot you constructed in exercise 1 of the "Data Visualization" exercises (average sighting latitude by year), this time adding a title, custom x and y-axis labels, and changing the color of the line to a hex color of your choosing. You should also customize the ticks on the y-axis to extend from 0 to 60 in increments of 10.

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings['latitude'] = pd.to_numeric(sightings['latitude'],errors='coerce')
```

```python
sightings['year'] = sightings['datetime'].dt.year
avg_latitude_by_year = sightings.groupby('year')['latitude'].mean()
```

```python
line_plot = avg_latitude_by_year.plot(color='#03e3fc', title='UFO Sighting Latitude By Year', xlabel='Sighting Year', ylabel = 'Average Sighting Latitude')
line_plot.set_yticks([0, 10, 20, 30, 40, 50, 60]);
```

## Recreate the bar plot you constructed in exercise 1 of the "Bar Plot" exercises (count of sightings by month), this time adding a title, custom x and y-axis labels, and changing the color of the bars to a hex color of your choosing. You should also customize the tick *labels* on the **x**-axis to reflect the actual month names, as opposed to their corresponding numbers.

```python
sightings['month'] = sightings['datetime'].dt.month
num_sightings_by_month = sightings['month'].value_counts().sort_index()
```

```python
bar_plot = num_sightings_by_month.plot(kind='bar', color='#0388fc', title='UFO Sightings By Month', xlabel='Sighting Month', ylabel = '# Sightings')
bar_plot.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']);
```
