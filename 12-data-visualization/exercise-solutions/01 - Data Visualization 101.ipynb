## Setup

```python
import pandas as pd
%matplotlib inline
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Plot average latitude by year:

```python
sightings['datetime'] = pd.to_datetime(sightings['datetime'],errors='coerce')
sightings['latitude'] = pd.to_numeric(sightings['latitude'],errors='coerce')
```

```python
sightings['year'] = sightings['datetime'].dt.year
avg_latitude_by_year = sightings.groupby('year')['latitude'].mean()
avg_latitude_by_year.plot()
```

## Plot the total number/count of sightings by year:

```python
num_sightings_by_year = sightings['year'].value_counts().sort_index()
num_sightings_by_year.plot()
```

## Can you identify any apparent relationship between average sighting latitude and total sighting count over the years? If so, can you think of a possible explanation for this relationship?

**It would appear that - in general - as the number of reported UFO sightings has increased significantly over the years, the trend in the average latitude of those sightings has become much more consistent, with fewer dramatic spikes or troughs.**

**This makes sense, because there tends to be more variance/"waviness" in small samples of data than in large ones. It is also easier for outliers to skew averages in smaller datasets.**

## Plot the total number/count of sightings by month:

```python
sightings['month'] = sightings['datetime'].dt.month
num_sightings_by_month = sightings['month'].value_counts().sort_index()
num_sightings_by_month.plot()
```

## In your assessment, are the results really a time series anymore, or more of a categorical breakout?

**These results don't reflect a linear passage of time anymore; after all, since our dataset spans many years, some sightings in January could have taken place long AFTER other sightings in (for example) April.**

**This, along with the relatively small number of discrete values on the x-axis, makes this more of a categorical breakout.**
