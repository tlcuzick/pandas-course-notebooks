## Setup

```python
import pandas as pd
%matplotlib inline
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Use a scatter plot to visualize the relationship between sighting *duration* and *latitude*.

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
sightings.plot(kind='scatter',x='duration (seconds)',y='latitude')
```

## Is there any apparent correlation between these variables? Can you identify any apparent clusters or outliers from looking at the plot?

**While most sightings seem to have a similarly short duration, regardless of latitude, the handful of sightings with longer durations generally seem to correlate with latitudes between ~30 and ~50 degrees latitude.**

**There also appears to be a sighting that occurred below -80 degrees latitude - far further south than the other sightings in the dataset - making it an "outlier" in that sense.**

## Use a scatter plot to visualize the relationship between sighting *latitude* and *longitude*.

```python
sightings.plot(kind='scatter',x='longitude',y='latitude')
```

## Can you identify any interesting or recognizable patterns in the resulting visualization?

**It turns out that when you plot the relationship between latitude and longitude on a scatter plot - especially if you have enough data points - you end up with an approximation of a map!**

**Which makes sense, given that we would only expect to see dots (representing sightings) on the landmasses where people actually live. Of course, parts of the world with fewer sightings will have a less defined/"filled in" map.**
