```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

```python
mask = sightings['comments'].str.contains('UFO',na=False,case=False)
sightings[mask].head(10)
```

```python
regex = '\d{2}/\d{2}/\d{4}'
contains_date = sightings['comments'].str.contains(regex,na=False,regex=True)
sightings[contains_date]
```

```python
contains_date = (sightings['comments'].str.contains('\d{2}/\d{2}/\d{4}',na=False,regex=True)) | (sightings['comments'].str.contains('\d{1}/\d{2}/\d{4}',na=False,regex=True))
sightings[contains_date].tail(10)
```
