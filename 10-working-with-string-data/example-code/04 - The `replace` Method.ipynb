```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head(10)
```

```python
sightings['shape'] = sightings['shape'].str.replace('sphere','ball')
sightings.head(10)
```

```python
mask = sightings['comments'].str.contains('unidentified flying object',na=False,case=False)
sightings[mask]
```

```python
sightings['comments'] = sightings['comments'].str.replace('unidentified flying object','UFO',regex=True,case=False)
```

```python
sightings[mask]
```

```python
sightings['comments'] = sightings['comments'].str.replace('&#44',',',regex=True)
sightings[mask]
```

```python
regex = '&#\d{2}'
```

```python
mask = sightings['comments'].str.contains(regex,na=False,regex=True)
sightings[mask]
```

```python
sightings['comments'] = sightings['comments'].str.replace(regex,' ',regex=True)
```

```python
sightings[mask]
```
