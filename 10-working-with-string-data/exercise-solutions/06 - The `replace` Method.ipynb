## In the "city" column of the UFO sightings DataFrame, replace any instances of the text strings "ft." and "mt.", with "fort " and "mount", respectively. Your solution should be case insensitive.

```python
mt = sightings['city'].str.contains('mt.', na=False, case=False, regex=False)
ft = sightings['city'].str.contains('ft.', na=False, case=False, regex=False)
sightings[mt | ft].head()
```

```python
cities = sightings['city'].str.replace('ft\.', 'fort', case=False, regex=True)
cities = cities.str.replace('mt\.', 'mount', case=False, regex=True)
sightings['city'] = cities
sightings[mt | ft].head()
```

## In the "city" column of the UFO sightings DataFrame, replace any text inside parentheses (including the parentheses) with an empty string:

```python
pattern = '\\(.*\\)'
```

```python
mask = sightings['city'].str.contains(pattern, na=False, regex=True)
sightings[mask].head()
```

```python
sightings['city'] = sightings['city'].str.replace(pattern, '', regex=True)
sightings[mask].head()
```
