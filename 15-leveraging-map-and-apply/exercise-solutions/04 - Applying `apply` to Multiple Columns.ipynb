## Create a custom "region" field in the UFO sightings dataset - based on the latitude and longitude of each respective sighting - using the below logic as a guide:

### North America
* Latitude: 0° to 90°
* Longitude: -180° to -60°

### Latin America
* Latitude: -55° to 23.5°
* Longitude: -120° to -35°

### Europe
* Latitude: 30° to 70°
* Longitude: -25° to 60°

### Africa
* Latitude: -34.5° to 23.5°
* Longitude: -20° to 40°

### Middle East
* Latitude: 15° to 40°
* Longitude: 30° to 60°

### Asia
* Latitude: 0° to 60°
* Longitude: 60° to 180°

### Oceania
* Latitude: 0°S to -60°S
* Longitude: 110°E to 180°E

## Other
* Any location not matching the above criteria

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
```

```python
def classify_region(row):
  latitude = row['latitude']
  longitude = row['longitude']

  if latitude >= 0 and latitude <= 90 and longitude >= -180 and longitude <= -60:
    return 'North America'
  elif latitude >= -55 and latitude <= 23.5 and longitude >= -120 and longitude <= -35:
     return 'Latin America'
  elif latitude >= 30 and latitude <= 70 and longitude >= -25 and longitude <= 60:
    return 'Europe'
  elif latitude >= -34.5 and latitude <= 23.5 and longitude >= -20 and longitude <= 40:
     return 'Africa'
  elif latitude >= 15 and latitude <= 40 and longitude >= 30 and longitude <= 60:
     return 'Middle East'
  elif latitude >= 0 and latitude <= 60 and longitude >= 60 and longitude <= 180:
     return 'Asia'
  elif latitude >= -60 and latitude <= 0 and longitude >= 110 and longitude <= 180:
    return 'Oceania'
  else:
    return 'Other'
```

```python
sightings['region'] = sightings.apply(classify_region, axis=1)
```
