```python
import pandas as pd
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

```python
def house_rater(row):
  price = row['price']
  bathrooms = row['bathrooms']
  grade = row['grade']
  condition = row['condition']

  if price >= 400000:
    return 'Too expensive!'
  elif (bathrooms - int(bathrooms)) in(0.25, 0.75):
    return 'Stupid number of bathrooms!'
  elif (grade > 7) and (condition > 3):
    return "I'm interested!"
  else:
    return "Not interested:("
```

```python
house_sales['my_rating'] = house_sales.apply(house_rater, axis=1)
```
