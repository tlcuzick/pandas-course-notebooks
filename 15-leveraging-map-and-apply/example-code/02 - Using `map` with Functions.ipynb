```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

```python
def classify_duration(duration):
  if duration >= 3600:
    return "very long"
  elif duration >= 600:
    return "long"
  elif duration >= 60:
    return "medium"
  elif duration >= 5:
    return "short"
  elif duration >= 0:
    return "very short"
  else:
    return None
```

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
duration_bucket = sightings['duration (seconds)'].map(classify_duration)
sightings.insert(6, 'duration bucket', duration_bucket)
sightings.head(10)
```

```python
regex = '&#\d{2}'
mask = sightings['comments'].str.contains(regex,na=False,regex=True)
sightings[mask]
```

```python
import html
```

```python
def convert_html(x):
  if type(x) == str:
    return html.unescape(x)
  else:
    return None
```

```python
sightings['comments'] = sightings['comments'].map(convert_html)
```
