## Given that there are the same number of possible degrees of latitude and longitude (360), what could explain the difference between the *standard deviation* of latitude and longitude?

**Just looking at the difference between the *max* and *min* latitudes and longitudes, it's clear that there is a much wider range of longitudes in which sightings occurred. This could easily lead to a higher *variance* in longitudes, hence the higher standard deviation.**

**One possible explanation is that latitude affects climate much more strongly than longitude, so exceptionally high or low latitudes are so sparsely populated that there simply aren't enough people to *have* many sightings in the first place.**
