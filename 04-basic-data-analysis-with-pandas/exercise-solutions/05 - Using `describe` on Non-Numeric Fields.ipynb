## Setup

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'YOUR PATH HERE'
ufo_sightings = pd.read_csv(ufos_path)
```

## Convert the "latitude" and "duration (seconds)" columns to numeric:

```python
ufo_sightings['duration (seconds)'] = pd.to_numeric(ufo_sightings['duration (seconds)'],errors='coerce')
ufo_sightings['latitude'] = pd.to_numeric(ufo_sightings['latitude'],errors='coerce')
```

## Apply the `describe` method to the DataFrame:

```python
ufo_sightings.describe(include='all')
```

## What state had the most sightings?

**California, abbreviated as "ca"**

## How many sightings occurred in that state?

**9,655**

## How many distinct countries are represented in the dataset?

**5**
