## Use the `value_counts` method to determine how many sightings are associated with the most recent "date posted" value in the UFO sightings dataset:

```python
dates_posted = pd.to_datetime(ufo_sightings['date posted'], errors='coerce')
dates_posted.value_counts().sort_index(ascending=False)
```
