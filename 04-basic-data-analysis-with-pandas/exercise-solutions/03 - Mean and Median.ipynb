## Setup

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'YOUR PATH HERE'
ufo_sightings = pd.read_csv(ufos_path)
```

## Convert the "latitude" and "duration (seconds)" columns to numeric:

```python
ufo_sightings['duration (seconds)'] = pd.to_numeric(ufo_sightings['duration (seconds)'],errors='coerce')
ufo_sightings['latitude'] = pd.to_numeric(ufo_sightings['latitude'],errors='coerce')
```

## Apply the `mean` and `median` methods to the DataFrame:

```python
ufo_sightings.mean(numeric_only=True)
```

```python
ufo_sightings.median(numeric_only=True)
```

## Interpret the difference between mean and median **sighting duration**:

**Mean sighting duration is MUCH longer than median sighting duration, implying that a small number of extremely long sightings is skewing the mean.**

## Interpret the difference (or lack thereof) between mean and median **latitude**:

**Mean and median latitude are very similar, implying that mean latitude isn't skewed by especially large (or small) values. This makes sense, given that there is a limited range of possible latitudes on planet Earth!**
