## Programatically calculate the number of values in the "datetime" column of the UFO sightings dataset that can't be converted to actual datetimes:

```python
ufos_path = 'YOUR PATH HERE'
ufo_sightings = pd.read_csv(ufos_path)
```

```python
datetimes_count = pd.to_datetime(ufo_sightings['datetime'],errors='coerce').count()
sightings_count = ufo_sightings.shape[0]
missing_datetime_values = sightings_count - datetimes_count
missing_datetime_values
```

**694 values in the "datetime" column of the UFO sightings dataset can't be converted to actual datetimes.**

## Use the `count` method to quality check the house sales dataset for missing values in any of the columns, or any values in the "date" column that can't be converted to a datetime data type:

```python
houses_path = 'YOUR PATH HERE'
house_sales = pd.read_csv(houses_path)
```

```python
house_sales.count()
```

**There are no missing values in any of the columns of the house sales dataset.**

```python
datetimes_count = pd.to_datetime(house_sales['date'],errors='coerce').count()
house_sales_count = house_sales.shape[0]
missing_datetime_values = house_sales_count - datetimes_count
missing_datetime_values
```

**All values in the "date" column can be converted to a `datetime` data type.**
