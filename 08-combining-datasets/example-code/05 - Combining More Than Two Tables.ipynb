```python
import pandas as pd
```

```python
customers_path = 'YOUR PATH HERE'
orders_2021_path = 'YOUR PATH HERE'
order_details_2021_path = 'YOUR PATH HERE'
```

```python
customers = pd.read_csv(customers_path)
orders_2021 = pd.read_csv(orders_2021_path)
order_details_2021 = pd.read_csv(order_details_2021_path)
```

```python
customer_orders_plus = pd.merge(orders_2021, order_details_2021, left_on='OrderID', right_on='OrderID', how='inner')
```

```python
customer_orders_plus = pd.merge(customers, customer_orders_plus, left_on='ID', right_on='CustomerID', how='left')
```
