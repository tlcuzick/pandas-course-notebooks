**It appears that it might be plausible to link the datasets together via the "ProductID" field in the Order Details dataset and the "ID" field in the products dataset, respectively.**

## What kind of relationship exists between the two datasets? What role does each dataset play in that relationship?

**It appears that the datasets are related via a *one-to-many* relationship, in which one product could potentially be associated with many order details/line items, but NOT the other way around.**

## Merge the columns of the two datasets together into a single consolidated dataset:

```python
product_orders = pd.merge(products,order_details_2021,left_on='ID',right_on='ProductID')
product_orders.head(10)
```

## Remove one of the fields used to connect the tables:

```python
product_orders.drop(labels='ID', axis=1, inplace=True)
product_orders.head(10)
```
