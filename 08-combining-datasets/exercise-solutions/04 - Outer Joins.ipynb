```python
import pandas as pd
```

```python
products_path = 'YOUR PATH HERE'
order_details_2021_path = 'YOUR PATH HERE'
order_details_2022_path = 'YOUR PATH HERE'
order_details_2023_path = 'YOUR PATH HERE'
```

## Generate a DataFrame consisting of all the columns from the "Products.csv" dataset combined with all the columns from the "Order Details 2021.csv" dataset. Only rows in which a match for a given product from "Products.csv" can be found in "Order Details 2021.csv" should be included.

```python
products = pd.read_csv(products_path)
order_details_2021 = pd.read_csv(order_details_2021_path)
```

```python
order_details_2021.head()
```

```python
product_sales_2021 = pd.merge(products, order_details_2021, left_on='ID', right_on='ProductID', how='left')
product_sales_2021.sort_values('OrderDetailID').head()
```

## Generate a DataFrame consisting of all rows from the "Products.csv" dataset associated with products that were not sold in 2022:

```python
order_details_2022 = pd.read_csv(order_details_2022_path)
```

```python
product_sales_2022 = pd.merge(products, order_details_2022, left_on='ID', right_on='ProductID', how='left')
```

```python
not_sold_2022 = product_sales_2022[product_sales_2022['OrderDetailID'].isna()]
not_sold_2022
```
