## Merge all the columns in the following datasets (provided as attachments in the resources for this exercise) into a single DataFrame:

* Customers.csv

* Orders 2022.csv

* Products.csv

* Order Details 2022.csv



```python
products_path = 'YOUR PATH HERE'
customers_path = 'YOUR PATH HERE'
orders_2022_path = 'YOUR PATH HERE'
order_details_2022_path = 'YOUR PATH HERE'
```

```python
products = pd.read_csv(products_path)
customers = pd.read_csv(customers_path)
orders_2022 = pd.read_csv(orders_2022_path)
order_details_2022 = pd.read_csv(order_details_2022_path)
```

```python
# preview order details dataset

order_details_2022.head()
```

```python
# preview orders dataset

orders_2022.head()
```

```python
# merge orders and order details datasets

order_info = pd.merge(orders_2022, order_details_2022, on='OrderID')
order_info.head()
```

```python
# preview products dataset

products.head()
```

```python
# add products dataset

orders_plus_products = pd.merge(products, order_info, left_on='ID', right_on='ProductID', how='left')
orders_plus_products.drop(labels='ProductID', axis=1, inplace=True) # drop redundant "ProductID" column
orders_plus_products.rename(columns={"ID": "ProductID"}, inplace=True) # rename remaining ID column to be more descriptive
orders_plus_products.head()
```

```python
# preview customers dataset

customers.head()
```

```python
# add customers dataset

all_in = pd.merge(customers, orders_plus_products, left_on='ID', right_on='CustomerID', how='left')
all_in.drop(labels='CustomerID', axis=1, inplace=True) # drop redundant "CustomerID" column
all_in.rename(columns={"ID": "CustomerID"}, inplace=True) # rename remaining ID column to be more descriptive
all_in.head()
```
