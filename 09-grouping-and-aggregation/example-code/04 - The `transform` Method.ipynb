```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(houses_path)
houses.head()
```

```python
houses.groupby('zipcode')['price'].mean()
```

```python
avg_by_zipcode = houses.groupby('zipcode')['price'].transform('mean')
avg_by_zipcode.get_group(98001)
```

```python
type(avg_by_zipcode)
```

```python
houses['avg_by_zipcode'] = avg_by_zipcode
houses[['zipcode','price','avg_by_zipcode']]
```
