```python
houses_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(houses_path)
houses.head()
```

```python
multi_level = houses.groupby(['bedrooms','bathrooms']).agg(avg_price=('price','mean'),max_price=('price','max'),avg_sqft=('sqft_living','mean'))
multi_level
```
