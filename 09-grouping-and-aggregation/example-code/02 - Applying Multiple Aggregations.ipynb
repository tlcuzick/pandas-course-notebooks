```python
houses_path = 'YOUR PATH HERE'
```

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
houses = pd.read_csv(houses_path)
houses.head()
```

```python
zipcode_stats = houses.groupby('zipcode')[['price','bedrooms','bathrooms']].agg(['mean','max','min'])
zipcode_stats
```

```python
zipcode_stats[('price', 'max')]
```

```python
zipcode_stats = houses.groupby('zipcode').agg(min_price=('price','min'), max_bedrooms=('bedrooms','max'), avg_bathrooms=('bathrooms', 'mean'))
zipcode_stats
```
