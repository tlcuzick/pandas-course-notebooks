```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(houses_path)
```

```python
houses.groupby(['bedrooms','bathrooms'])['price'].mean()
```

```python
houses.pivot_table(index='bathrooms', columns='bedrooms', values='price', aggfunc='mean')
```

```python
houses.pivot_table(index=['waterfront','view'], columns='condition', values='price', aggfunc='mean')
```

```python
houses.pivot_table(index=['waterfront','view'], columns='condition', values='price', aggfunc=['mean','max'])
```

```python
houses.pivot_table(index='yr_built', values=['price','bedrooms'], aggfunc='mean')
```
