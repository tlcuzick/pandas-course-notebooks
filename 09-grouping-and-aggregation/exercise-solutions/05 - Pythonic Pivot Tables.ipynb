## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Break out the `mean` of sighting duration in **minutes** by "country" (on columns) and "shape" (on rows):

```python
sightings['duration (minutes)'] = pd.to_numeric(sightings['duration (seconds)'],errors='coerce') / 60
```

```python
sightings.pivot_table(index='shape',columns='country',values='duration (minutes)',aggfunc='mean')
```

## Modify your solution to the previous exercise to include a `count` of values in your new "duration (minutes)" column as an additional aggregate statistic:

```python
sightings.pivot_table(index='shape',columns='country',values='duration (minutes)',aggfunc=['count','mean'])
```

## Break out the `mean` of "duration (minutes)" AND the `mean` of "latitude", by "country" (on columns) and "shape" (on rows):

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'],errors='coerce')
```

```python
sightings.pivot_table(index='shape',columns='country',values=['duration (minutes)','latitude'],aggfunc='mean')
```

## Break out the `count` of "date posted" by "country" AND "state" on rows, and by "shape" on columns:

```python
sightings.pivot_table(index=['country','state'],columns='shape',values='duration (minutes)',aggfunc='count')
```

## Make the layout of your solution to the previous exercise more manageable by filtering the UFO sightings dataset for sightings with the following shapes: "changed", "changing", "chevron", or "cigar":

```python
shapes = ['shape',	'changed',	'changing',	'chevron',	'cigar']
sightings[sightings['shape'].isin(shapes)].pivot_table(index=['country','state'],columns='shape',values='duration (minutes)',aggfunc='count')
```
