```python
sightings = pd.read_csv(ufos_path)
sightings.head()
```

## Using the `groupby` method, break out the count of values from ALL fields in the UFO sightings dataset by **country**.

```python
sightings.groupby('country').count()
```

 For which country are the count of values for "city" and "state" identical?

**The United States**

## Break out the average of "duration (minutes)" by **country**.

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
sightings.groupby('country')['duration (seconds)'].mean()
```

Which country had the shortest average sighting duration?

**Australia**

## Break out the average of "duration (minutes)" AND "latitude" by sighting **shape**.

```python
sightings['latitude'] = pd.to_numeric(sightings['latitude'], errors='coerce')
sightings.groupby('shape')[['duration (seconds)', 'latitude']].mean()
```

## Using the `loc` method, limit the results of the previous exercise to just the following shapes: "pyramid", "hexagon", "triangle", "rectangle", and "diamond".

```python
gp = sightings.groupby('shape')[['duration (seconds)', 'latitude']].mean()
gp.loc[['pyramid', 'hexagon', 'triangle', 'rectangle', 'diamond']]
```

Of sightings with these shapes, which had the **highest** average latitude? The **lowest** average duration?

**The "pyramid" shape for both!**
