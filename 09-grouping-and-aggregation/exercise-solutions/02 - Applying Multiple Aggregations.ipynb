## Break out the `mean`, `min`, `max`, and `count` of "duration (seconds)" by "state":

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
sightings.groupby('state')['duration (seconds)'].agg(['mean','min','max','count'])
```

## Break out the following aggregate statistics by "state":

* The `mean` of "sighting duration (seconds)", in a new column called "mean_duration"

* The `max` of "sighting duration (seconds)", in a new column called "max_duration"

* The `count` of "shape", in a new column called "recorded_shape_ct"

Your solution should **not** have a multi-index.

```python
sightings.groupby('state').agg(mean_duration=('duration (seconds)','mean'),max_duration=('duration (seconds)','max'),recorded_shape_ct=('shape','count'))
```
