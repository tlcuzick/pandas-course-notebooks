## Setup

```python
import pandas as pd
```

```python
house_path = 'YOUR PATH HERE'
```

```python
houses = pd.read_csv(house_path)
houses.head()
```

## In the house sales dataset, create a new derived column called "avg_beds_baths_per_zip" that displays the average/mean number of combined bedrooms + bathrooms for the zip code in a given row:

```python
houses['beds_plus_baths'] = houses['bedrooms'] + houses['bathrooms']
houses['avg_beds_baths_per_zip'] = houses.groupby('zipcode')['beds_plus_baths'].transform('mean')
```

```python
houses[['zipcode','bedrooms','bathrooms','beds_plus_baths','avg_beds_baths_per_zip']].sort_values('zipcode')
```

## Add another derived column to the house sales dataset that displays the difference between the sale price in a given row, to the average price of other homes built in the same year. Name this column "age_cohort_price_delta".

```python
age_cohort_avg_price = houses.groupby('yr_built')['price'].transform('mean')
houses['age_cohort_price_delta'] = houses['price'] - age_cohort_avg_price
```

```python
houses[['price','yr_built','age_cohort_price_delta']].sort_values('yr_built')
```
