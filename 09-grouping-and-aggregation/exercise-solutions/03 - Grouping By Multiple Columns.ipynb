## Break out the following aggregate statistics by "city", "state", and "country", in that order:
* The `mean` of "sighting duration (seconds)", in a new column called "mean_duration"

* The `max` of "sighting duration (seconds)", in a new column called "max_duration"

* The `count` of "shape", in a new column called "recorded_shape_ct"

```python
sightings['duration (seconds)'] = pd.to_numeric(sightings['duration (seconds)'], errors='coerce')
```

```python
gb_fields = ['city','state','country']
sightings.groupby(gb_fields).agg(mean_duration=('duration (seconds)','mean'),max_duration=('duration (seconds)','max'),recorded_shapes=('shape','count'))
```

## Does the order in which the groupby fields appear in the multi-index make sense? Why or why not?

**It does not!**

**If there are natural parent-child relationships between two or more fields you're grouping by (e.g., any one value in the "parent" field is typically associated with *multiple* values in the "child" field, but not the other way around), the "parent" fields should come first.**

## If not, *change* that order to make the output easier to interpret and analyze:

```python
gb_fields = ['country','state','city']
sightings.groupby(gb_fields).agg(mean_duration=('duration (seconds)','mean'),max_duration=('duration (seconds)','max'),recorded_shape_ct=('shape','count'))
```

## Modify your solution to the previous exercise by resetting the index of the resulting DataFrame. Display the results.
