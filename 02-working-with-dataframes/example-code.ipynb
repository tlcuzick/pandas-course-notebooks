# Renaming Columns

```python
import pandas as pd
```

```python
houses_path = 'YOUR PATH HERE'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

```python
new_df = house_sales.rename(columns={'date': 'sale_date', 'sqft_above': 'sqft_aboveground'})
new_df.columns
```

```python
house_sales.rename(columns={'date': 'sale_date', 'sqft_above': 'sqft_aboveground'},inplace=True)
house_sales.columns
```
