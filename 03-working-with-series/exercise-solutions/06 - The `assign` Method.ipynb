## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

## Define the following Series, and assign them to their own respective variables:

* A Series consisting of the "datetime" column converted to a datetime data type

* A Series consisting of the "duration (seconds)" column converted to a numeric data type

* A Series - with a numeric data type - consisting of the "duration (seconds)" column converted to minutes

* A Series consisting of the "date posted" column converted to a datetime data type

* A Series consisting of the "latitude" column converted to a numeric data type

```python
datetime=pd.to_datetime(ufo_sightings['datetime'],errors='coerce')
duration_seconds=pd.to_numeric(ufo_sightings['duration (seconds)'],errors='coerce')
duration_minutes=duration_seconds / 60
date_posted=pd.to_datetime(ufo_sightings['date posted'],errors='coerce')
latitude = pd.to_numeric(ufo_sightings['latitude'],errors='coerce')
```

## Add all of the above Series to a new copy of the original DataFrame - in a single line of code - using the `assign` method. Overwrite the original versions of the columns where possible/appropriate

```python
new_df = ufo_sightings.assign(datetime=datetime,duration_seconds=duration_seconds,duration_minutes=duration_minutes, date_posted=date_posted, latitude=latitude)
new_df.info()
```

## Use the `drop` method to remove the original versions of the converted DataFrame columns that you were unable to overwrite with `assign`.

```python
new_df.drop(labels=['duration (seconds)', 'date posted'], axis=1, inplace=True)
new_df.head()
```
