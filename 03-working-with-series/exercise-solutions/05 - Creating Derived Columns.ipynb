## Setup

```python
import pandas as pd
```

```python
ufos_path = 'YOUR PATH HERE'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

## Add a "duration (minutes)" column to the DataFrame, immediately following the existing "duration (seconds)" column:

```python
ufo_sightings.insert(6, 'duration (minutes)', pd.to_numeric(ufo_sightings['duration (seconds)'], errors='coerce') / 60)
```

## Add a "distance from 37th" column, containing the difference between the latitude of a given sighting, and the 37th parallel:

```python
ufo_sightings['distance from 37th'] = abs(pd.to_numeric(ufo_sightings['latitude'], errors='coerce') - 37)
```
